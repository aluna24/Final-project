```python
class RPS(object):
   
    def validate(self,move):      
        while True:    
            move = input("Please input a move\n")
            if move == "rock" or move == "paper" or move == "scissors" or move == "ice" or move == "fire":
                return move
            else:
                print("Not a valid move! Please try again and enter a valid move.")
   
    def result(self,player1,player2):
        #while True:
        player1_score = 0
        player2_score = 0
        if player1 == "rock": #set the winning and losing conditions for rock
            if player2 == "paper" or player2 == "fire": # player2's paper/dog beats player1's rock
                winner = "Player 2" #setting up a variable to declare the winner
                wMove = player2 #setting up a variable to declare the winning move
                loser = "Player 1" #variable for loser
                lMove = player1 #variable for losing move
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player2_score += 1
            elif player2 == "scissors" or player2 == "ice": #if player2 inputs scissors/cat, player 1 wins
                winner = "Player 1"
                wMove = player1
                loser = "Player 2"
                lMove = player2
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player1_score += 1
            else:
                print("It's a tie. Boring!")

        elif player1 == "paper": #then i set the conditions for paper following the same framework as rock
            if player2 == "scissors" or player2 == "ice": # scissors/cat beats paper
                winner = "Player 2"
                wMove = player2
                loser = "Player 1"
                lMove = player1
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player2_score += 1
            elif player2 == "rock" or player2 == "fire": #paper beats rock/dog
                winner = "Player 1"
                wMove = player1
                loser = "Player 2"
                lMove = player2
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player1_score += 1
            else:
                print("It's a tie. Boring!")

        elif player1 == "scissors": #same w scissors
            if player2 == "rock" or player2 == "fire": #rock/dog beats scissors
                winner = "Player 2"
                wMove = player2
                loser = "Player 1"
                lMove = player1
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player2_score += 1
            elif player2 == "paper" or player2 == "ice": #paper/cat loses to scissors
                winner = "Player 1"
                wMove = player1
                loser = "Player 2"
                lMove = player2
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove) #then we print the winning statement
                player1_score += 1
            else:
                print("It's a tie. Boring!")
       
        elif player1 == "ice": #same w cat
            if player2 == "rock" or player2 == "scissors": #rock/scissors beats cat
                winner = "Player 2"
                wMove = player2
                loser = "Player 1"
                lMove = player1
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove)
                player2_score += 1
            elif player2 == "fire" or player2 == "paper": #cat beats dog and paper
                winner = "Player 1"
                wMove = player1
                loser = "Player 2"
                lMove = player2
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove)
                player1_score += 1
            else:
                print("It's a tie. Boring!")
       
        elif player1 == "fire": #finally for dog
            if player2 == "ice" or player2 == "paper": #cat/paper beats dog
                winner = "Player 2"
                wMove = player2
                loser = "Player 1"
                lMove = player1
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove)
                player2_score += 1
            elif player2 == "scissors" or player2 == "rock": #dog beats scissors/rock
                winner = "Player 1"
                wMove = player1
                loser = "Player 2"
                lMove = player2
                print(winner+"'s "+wMove+" beats "+loser+"'s "+lMove)
                player1_score += 1
            else:
                print("It's a tie. Boring!")
               
        return player1_score,player2_score
               
    def scorekeeper(self,p1_score,p2_score):
        p1 = 0
        p2 = 0
        if p1_score == 1:
            p1 += 1
        elif p2_score == 1:
            p2 += 1
        return p1, p2
   
    def run_game(self):

        x = 0
        y = 0
        p1 = 0
        p2 = 0
        k = 0
        s = 0
        while True:

            a = self.validate(self)

            b = self.validate(self)

            p1,p2 = self.result(a,b)
            x,y = self.scorekeeper(p1,p2)
               
            if x == 1:
                k = k + x
                   
            elif y == 1:
                s = s + y

            print("Player1 score:", k)
            print("Player2 score:", s)
               
            if k == 2 or s ==2:
                break    
        if k == 2:
            print("Player 1 wins!")     
        else:
            print("Player 2 wins!")
        return 1
       
def main():
    print("Let's play rock, paper, scissors, ice, fire") #setting up the opening statement
    game = RPS()
    while True:
        g = game.run_game()
        if g == 1:
            break

if __name__ == "__main__":
    main()
```
